```python
#делаем колонку с преобразованием даты в количество месяцев до вопроса
#и колонку с номером месяца
# для train
new_column = np.array([])
itog_month = np.array([])
for cur_id in train['SK_ID'].unique():
    years = []
    months = []
    days = []
    count_months = []
    cur_train = train[train['SK_ID'] == cur_id]
    for cur in cur_train["SNAP_DATE"]:
        d,m,y = map(int, str(cur).split('.'))
        years.append(y)
        months.append(m)
        days.append(d)
        count_months.append(m + 12 * (y - 1))
    count_months = np.array(count_months)
    max_count_monhts = count_months.max()
    raz_month = count_months * (-1) + max_count_monhts
    
    if (len(new_column) != 0):
        new_column = np.concatenate((new_column, raz_month), axis = 0)
        itog_month = np.concatenate((itog_month, months), axis = 0)
    else:
        new_column = raz_month
        itog_month = months
train['date_dist_from_contact'] = new_column
#train['month'] = itog_month  
```

```python
import copy
```

```python
#преобразуем признаки(здесь я учитываю за 2,1,0 месяцев до опроса)

def new(df):
    new_df = pd.DataFrame()
    names_column = []
    for name in dif_column:
        for i in range(0, 4):
            names_column.append(str(i) + name)
            new_df[str(i) + name] = [np.float32(-1.0)]
    new_df['SK_ID'] = [int(-1)]
    for cur_id in df['SK_ID'].unique():
        part_df = df[df['SK_ID'] == cur_id]
        new_line = copy.copy(new_df.loc[0])
        for dist_date in range(0, 4):
            for column in dif_column:
                
                if dist_date in part_df['date_dist_from_contact'].values:
                    new_line[str(dist_date) + column] = \
                    part_df[part_df['date_dist_from_contact'] == dist_date][column]
                else:
                    new_line[str(dist_date) + column] = -2.0
        new_line['SK_ID'] = cur_id
        new_df.loc[np.shape(new_df)[0]] = new_line     
        
        
    return new_df
```
