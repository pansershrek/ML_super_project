```python
import numpy as np
import pandas as pd
import re
```

```python
with open('bs_avg_kpi_butch', 'w') as f:
    ind = 0
    with open('bs_avg_kpi.csv', 'r') as file:
        for line in file:
            print(line.strip(), file = f, flush=True)
            ind += 1
            if ind > 10:
                break
```

```python
tmp = pd.read_csv('bs_avg_kpi_butch', sep=';')
```

```python
cmd = 'create table bs_avg_kpi('
for col in range(len(tmp.columns)):
    if col > 0 and col < len(tmp.columns):
        cmd = f'{cmd},'
    if tmp[tmp.columns[col]].dtype == np.int64:
        cmd = f'{cmd} {tmp.columns[col]} bigint'
    if tmp[tmp.columns[col]].dtype == np.float64:
        cmd = f'{cmd} {tmp.columns[col]} float8'
    else:
        cmd = f'{cmd} {tmp.columns[col]} varchar'
cmd = f'{cmd});'
print(cmd)
```

```python
with open('bs_avg_kpi.csv', 'r') as file:
    with open('bs_abg_kpi_tmp', 'w') as file1:
        ind = 0
        for line in file:
            ind += 1
            tmp = re.sub(r',' , '0.', line)
            tmp = re.sub(r';' , ',' , tmp)
            print(tmp.strip(), file=file1, flush=True)
            if ind % 10000 == 0:
                print(ind, line)
```
